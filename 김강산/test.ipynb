## 압축해제 에러난 파일 다시 압축해제

```python
cd /content/drive/MyDrive/영양4조/dataset/음식 이미지 및 영양정보 텍스트/Train/txt
```

```python
folder_list = glob.glob('*/')
print(len(folder_list))
folder_list
```

```python
empty_folder_list = []
for folder in tqdm(folder_list):
  file_list = glob.glob(folder+'*')
  if len(file_list) < 1000: 
    empty_folder_list.append(folder)
```

```python
empty_folder_list
for idx, folder in enumerate(empty_folder_list):
  empty_folder_list[idx] = folder.split('/')[0] + '.zip'
```

```python
print(empty_folder_list)
print(len(empty_folder_list))
```

```python
for i in tqdm(empty_folder_list):
  os.system('mkdir '+unicodedata.normalize('NFC', i.split('.')[0]))
  os.system('unzip -qq ' + unicodedata.normalize('NFC', i) + ' -d ./{}'.format(unicodedata.normalize('NFC', i.split('.')[0])))
  print(i, '해제')
```
