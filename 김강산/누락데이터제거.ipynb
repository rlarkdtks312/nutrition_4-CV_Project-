```python
import glob
import os
import sys
from tqdm.notebook import tqdm
```

```python
class_list = glob.glob('/content/drive/MyDrive/영양4조/dataset/음식 이미지 및 영양정보 텍스트/Train/train/*')
len(class_list)
```

```python
zero_list = []
for cl in tqdm(class_list):
  img_list = glob.glob(cl+'/*.*')
  if len(img_list) == 0:
    zero_list.append(cl)
```

```python
zero_list
```

```python
for path in tqdm(zero_list):
  os.system('rm -r {}'.format(path))
```
