```python
# zip파일 압축해제하기
# 분할 압축되었던 zip 파일 해제하기
for zip in tqdm(zip_list):
  os.system('zip -F {} --out merged.zip'.format(zip))
  print(zip, "분할 압축 파일 merged")
  os.system('unzip merged.zip')
  print(zip, "파일 압축해제 완료")
  os.system('rm -r merged.zip')
  print(zip, "merged 파일 제거완료")
```
